### Subplots

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, "b")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
plt.show()
```

```python
fig, axes = plt.subplots(2,4) #2 filas y 4 columnas del lienzo

#genera un trazo accediendo a las graficas a traves del indice de la matriz
axes[0,0].plot(x,y)
axes[0,1].plot(y,x, 'r')

fig.tight_layout() #mejora la visualización de los ejes de cada gráfico
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)) = plt.subplots(2,4)
ax1.plot(x, np.sin(x), 'b')
ax2.plot(y, np.cos(x), 'purple')
ax3.plot(x, np.tan(y), 'orange')
ax4.plot(y, np.cos(y)**2, 'black')

#mejorar la visualización de las figuras
fig.tight_layout()
```

### Leyendas, etiquetas, títulos, tamaño

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (5,5))
ax1.plot(x,y, 'b', label ="$sin(x)$")
ax1.set_title("Relacion X - Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
ax2.plot(y,x, 'r')
ax2.set_title("Relacion Y - X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
plt.show()
```

```python
plt.plot(x,y, label = "sin(x)")
plt.title("título de la figura")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc = "upper right")
```

### Bar plot
El gráfico de barras nos permite graficar variables categóricas, es decir, variables de texto, lo que es muy importante en el mundo de la ciencia de datos y Matplotlib ofrece ciertas características que nos facilita la vida en cuanto a graficar este tipo de variables.

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [5000, 800, 900, 1000, 300]
```

```python
plt.bar(countrys,population)
plt.show()
```

```python
plt.bar(countrys,population,  width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.show()
```

```python
#Modificar los nombres de las variables en el eje ‘X’ con xticks
plt.bar(countrys,population, width=0.5, color= ["aqua", "grey", "teal", "crimson", "violet"])
plt.xticks(np.arange(5), ('India','Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

```python
#gráfico de barras horizontales
plt.barh(countrys,population)
plt.show()
```

Gracias a la librería GridSpec se puede combinar un gráfico scatter plot de 2 variables con los histogramas de cada una de las 2 variables.
Con esta librería definimos una figura de 4x4 gráficos. Luego compartimos unimos zonas de la cuadrícula de gráficos definidos de manera que:
Para el scatter plot quede una zona de 3x3, para el histograma de x una zona de 1x3 y para el histograma de y una zona de 3x1.


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig = plt.figure()
gs = GridSpec(4, 4)

ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_y = fig.add_subplot(gs[0,0:3])
ax_hist_x = fig.add_subplot(gs[1:4, 3])

plt.show()
```

```python
fig = plt.figure()
gs = GridSpec(4, 4)
ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_hist_y = fig.add_subplot(gs[0,0:3])
ax_hist_x = fig.add_subplot(gs[1:4, 3])

N = 50 
x = np.random.rand(N)
y = np.random.rand(N)

df = pd.DataFrame({'x': x, 'y': y})

ax_scatter.scatter(df.x,df.y)
ax_hist_x.hist(df.x, orientation = 'horizontal')
ax_hist_y.hist(df.y)

plt.show()
```

### Crear un gráfico tipo Scatter



```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area) # el tamaño de las burbujas hace referencia la área
```

```python
plt.scatter(x,y, s=area, c= colors, marker = 'p') # cambiar el marker
plt.show()
```

```python
plt.scatter(x,y, s=area, c= colors, marker = 'v', alpha= 0.5) # transparencia
plt.show()
```
