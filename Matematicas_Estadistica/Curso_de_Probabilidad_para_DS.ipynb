Sean A y B dos eventos aleatorios diferentes, entonces:

---

`Probabilidad conjunta (joint)`: sean $A$ y $B$ dos eventos aleatorios diferentes, definimos la probabilidad conjunta de que ambos eventos ocurran por medio de la regla del producto: 

$$P(A, B) = P(A|B)P(B)$$

donde $P(A|B)$ se define como la `probabilidad condicional` de que ocurra A dado B, y $P(B)$ es la probabilidad de que ocurra B (sin conocimiento extra de otras variables).

---

`Probabilidad marginal`: la probabilidad marginal de A sobre una distribucin conjunta de probabilidades en A y B, se define como la suma de todas las probabilidades conjuntas $P(A, B)$ considerando todos los posibles valores de B: 

$$P(A) = \sum_{B=b} P(A,B) = \sum_{B=b} P(A|B) P(B)$$

### Distribuciones de Probabilidad/ funcion densidad

Una Distribuciones de Probabilidad es una funcin que toma una variable aleatoria y a cada uno de sus posibles estados dentro del esacio muestral le asigna una probabilidad.

X  variable aleatoria

### Distribuciones de probabilidad

#### Distribuciones discretas

1. **Distribucin Binomial o de Bernoulli**

Este tipo de distribucin se obtiene a partir de un proceso llamado ensayo de Bernoulli. Este ltimo es un ensayo o experimento que puede conducir slo a uno de dos resultados mutuamente excluyentes, es decir nos permite describir ocurrencias con variables que son binarias por ejemplo en el caso de lanzar una moneda solo hay dos posibles resultados cara o cruz (otros ej. muerto o vivo, enfermo o sano, femenino o masculino). En una secuencia de ensayos o proceso de Bernoulli, la distribucin binomial nos permite calcular la probabilidad de obtener un resultado cualquiera (por ej. k xitos), para un nmero n de pruebas independientes, para ello se deben cumplir las siguientes condiciones:

1. Cada ensayo conduce a uno de dos resultados posibles mutuamente excluyentes. Uno de ellos se denota arbitrariamente como xito (x=1) y el otro como fracaso (x=0)
2. La probabilidad de xito, denotada por p, permanece constante de ensayo a ensayo. La probabilidad de fracaso es 1-p.
3. Los ensayos son independientes, es decir, que el resultado de cualquier ensayo particular no se ve afectado por el resultado de otro ensayo.




$$ 
     \begin{cases}
        P(X=1) = p \\
        P(X=0) = 1-p \\
     \end{cases}
$$


Recordando que la distribucin binomial est dada por: 

$$
P(k, n; p) = {n \choose k} p^k (1-p)^{n-k} = \frac{n!}{k!(n-k)!}p^k (1-p)^{n-k}
$$

donde $P(k, n; p)$ representa la probabilidad de obtener $k$ xitos de $n$ intentos con posibilidad **binaria** (por ejemplo, lanzamientos de moneda).

#### Distribucin normal terica (Gaussiana)


$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

* $\mu$: media de la distribucin
* $\sigma$: desviacin estndar de la distribucin

`Ejemplo`: considerando una variable aleatoria que sigue una distribucin normal con media $\mu=4$ y desviacin estndar $\sigma=0.3$, la probabilidad de que dicha variable tome el valor de $0.2$ est dada por: 

$$P(0.2) =  \frac{1}{0.3 \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{0.2-4}{0.3} \right)^2 \right]}$$

Lo cual en Python se traduce en: 

```python
from scipy.stats import norm

norm(mu, sigma).pdf(X)
```

---

As tambin, la distribucion de probabilidad acumulada correspondiente est dada por: 

$$C(X) = \int_{x<=X} P(x)dx = \int_{- \infty}^xP(X)dX $$

teniendo en cuenta que $Dom(X) = (-\infty, \infty)$. 

`Ejemplo`: considerando una variable aleatoria que sigue una distribucion normal con media $\mu=4$ y desviacin estndar $\sigma=0.3$, la probabilidad de que dicha variable tome el valor de $0.2$ o menos est dada por:

$$C(0.2) = \int_{x<=0.2} P(x)dx = \int_{- \infty}^{0.2}  \left\{\frac{1}{0.3 \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-4}{0.3} \right)^2 \right]} \right\} dX $$

La cual se calcula en Python como: 

```python 
from scipy.stats import norm

norm(mu, sigma).cdf(X)
```

_**Es importante recordar que la funcin de probabilidad acumulada de la distribucin gaussiana no se puede calcular de forma exacta, ya que la integral anterior no tiene una expresin cerrada conocida. Es decir, los mtodos de integracin conocidos no funcionan para resolver esta integral.**_

### Como estimar una Distribucion?

La filosofia del Machine Learning es ajustar una distribucion a un conjunto de datos para que con dicha distribucion podamos hacer predicciones

Tenemos dos tipos de estimaciones:

1. Estimacion Parametrica: Se basa en forzar los parametros de la distribucion para ajustarla a la distribucion de los datos en estudio

2. Estimacion no Parametrica: no se fuerzan los parametros de una distribucion unica, sino que se hace una combinacion de varias distribuciones. Depende de otros parametros como el parametro de suavizado y el tipo de funcion base utilizada para el ajuste

En Machine Learning y en general en todo modelo que se haga sobre un conjunto de datos, consiste en que siempre estas intentando ajustar una densidad de probabilidad a un conjunto de datos real

### Estimacin de mxima verosimilitud MLE

Nos permite estimar densidades de probabilidad en un esquema de trabajo muy general

Elementos esenciales:
1. elegir una distribucin. Teniendo solo una muestra de los datos

2. Elegir los parmetros de la distribucin. Para ajustar mejor la distribucin a los datos

En general como contamos con una muestra de la poblacin general de datos la distribucin de los datos de la muestra difiere de la distribucin de los datos de la poblacin 

MLE (Maximum Likelihood Estimation, o estimacin de mxima verosimilitud. ) es un problema de optimizacin

Consiste en escoger la distribucion y los parametros que ms se ajuste a nuestros datos, para ello maximizamos la probabilidad de nuesto modelo (mximo de la funcin L):
P(X,) = L(X,)

max L(X,)  maxP(Xi,)

Nota:
Cuando se factorizan las probabilidades como el producto de mtiples probabilidades, se generan nmeros cada vez ms pequeos al punto que no son reconocidos por la computadora, a esto se llama underflow para evitar esta situacion usamos logaritmos de las probabilidades:
Existe una propiedad de que el logaritmo de un producto es la suma de los logaritmos, esto convierte el problema de producto a sumas, los convierte en nros negativos ms grandes que son mas faciles de computar

max log L(X,)  max log P(Xi,)
