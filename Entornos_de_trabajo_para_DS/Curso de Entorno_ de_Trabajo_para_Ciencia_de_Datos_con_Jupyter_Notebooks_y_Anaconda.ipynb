Lo ideal es tener un ambiente para cada proyecto que tenga sus dependencias propias

**Entorno de trabajo:** Proyecto que puede tener sus dependencias propias, independientemente de las dependencias que tengas en los demás proyectos

**Anaconda:** programa multiplataforma que se utiliza para la gestión de paquetes, dependencias y entornos para cualquier lenguaje, por ej python, R, etc

### Anaconda
    
conda info        
    me da informacion    

conda env list            
    me lista los environments

conda create --name py35 python=3.5 pandas           
    Me permite crear un entorno el nombre del entorno es py35 y especifico la version de python que quiero instalar, en este caso 3.5, de pandas me va a instalar la que haya disponible más actualizada si no especifico la versión

conda list       
    me lista todo lo que está instalado

conda update pandas    
    me permite actualizar una libreria a su versión mas reciente, en este caso pandas

conda install pandas=1.2           
    con este comando puedo especificar el número de versión que quiero instalar de la libería, en este caso de pandas
    
conda create --name py39 --copy --clone py35
    de esta forma creo un nuevo entorno llamado py39 y clono los programas y librerias que hay en el otro entorno que tengo creado que es py35
    
##eliminar librerías y entornos
    
conda env list
    listo los entornos que tengo el * me indica en el entorno que estoy ahora

conda remove pandas 
    borro pandas del entorno donde estoy
conda list pandas
    chequeo si se eliminó
    
conda env remove --name py35
    elimino el entorno py35
conda environments
    chequeo si se eliminó
    
Nota: no puedo eliminar un entorno si está activado, para eso tendría que poner 
    conda deactivate py39
Si es que quisoera eliminar el entorno py39 y estudiera activado.
    

### instalación de paquetes 
    por ejemplo boltons
    
conda install --channel conda-forge boltons
    utilizo esto cuando con conda install y el nombre del paquete me sale que no lo encuentra, voy a la pagina que especifica conda
    
conda list --revision 
    me da una lista de los cambios que hice en mi entorno, esto me sirve para volver a una revisión anterior por ejemplo a alguna en donde boltons no estuviera instalada. En este caso la revisión que incluye boltons es la 1
    
si quiero volver a la revisión anterior
conda install --revision 0
    
##exportar un ambiente en forma de archivo
    
conda env --from-history --file environment.yml
ls      verifico el archivo creado .yml
cat environment.yml  me da una lista de las versiones que contiene el archivo
    
si quisiera replicar este ambiente en mi computadora, suponiendo que es de otra persona que me lo comparte
conda env create --file environment.yml
    
Para acelerar la descarga de paquetes y creación de ambientes de trabajo puedo utilizar mamba en lugar de conda
conda install --channel conda-forge mamba
    puedo hacer lo mismo que conda pero poniendo mamba al principio en lugar de conda
 
    
### Bonus: Divide y vencerás.

En proyectos grandes, es complicado mantener un ambiente virtual. 😵
Divide y vencerás es dividir un problema difícil en partes pequeñas. Cada paquete se desarrolla a su propio ritmo (no siempre preocupándose de la compatibilidad con otros módulos). 🤯
Hay tres tipos de paquetes: Externos, Modelo y de Comunicación. Entonces podemos crear en multiples ambientes para un solo proyecto. 👀 Para esto, creamos una carpeta llamada envs y creamos tres documentos: external.yml, model.yml y comunicacion.yml.
Snakemake es un paquete que implementa muy bien este concepto. Es un motor de workflows 👀. Cada paso lo ejecuta con un ambiente específico.
https://snakemake.readthedocs.io/en/stable/
