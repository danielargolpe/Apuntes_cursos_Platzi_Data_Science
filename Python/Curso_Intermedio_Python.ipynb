# filter, map, reduce

Hay tres funciones de orden superior más importantes en los lenguajes de programación: filter, map y reduce


```python
#aplicando filter
#El ejercicio anterior puede resolverse con filter, que es una función de orden superior que va a recibir 2
#parámetros una función anónima y un iterable (que es este caso es my_list)

my_list = [1, 4, 5, 6, 9, 13, 19, 21]
odd = (list(filter(lambda x: x % 2 != 0, my_list))) 
print(odd)

#la función filter por sí misma no nos va a devolver la lista que nosotros queremos porque retorna un iterador, 
#esto lo soluciono envolviendo la función en otra función list 
```

```python
my_list = [1, 2, 3, 4, 5]
squares = [i **2 for i in my_list] #quiero obtener una nueva lista del cuadrado de los valores
print(squares)
```

```python
#aplicando map
my_list = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x **2, my_list)) 
print(squares)

#la función map que es una función de orden superior que va a recibir 2
#parámetros una función anónima y un iterable (que es este caso es my_list)
# al igual que la función filter, map por sí misma no nos va a devolver la lista que nosotros queremos porque retorna un iterador, 
#esto lo soluciono envolviendo la función en otra función list 
```

```python
my_list = [2, 2, 2, 2, 2]
all_multiplied = 1
for i in my_list:
    all_multiplied = all_multiplied * i
print(all_multiplied)
```

```python
#aplicando reduce
from functools import reduce #debo importarla
my_list = [2, 2, 2, 2, 2]
all_multiplied = reduce(lambda a, b: a * b, my_list)
print(all_multiplied)

#la función reduce retorna un elemento único
```

### Manejo de errores 


Errores en el código
Cuando python nos avisa que tenemos un error en el código nos da un mensaje que conocemos como traceback, puede ser debido a:

Errores de Sintaxis (SyntaxError) → escribimos mal alguna palabra clave (typo), el programa no se ejecuta.
Excepciones (Exeption) → Producen un colapso o interrupción de la lógica del programa en alguna línea específica por ejemplo (todas las líneas anteriores se ejecutan), pueden ser de varios tipos, generalmente aparecen cuando no existe un componente clave en la ejecución o hay alguna imposibilidad lógica (matemática) para efectuar la instrucción, también pueden generarse dentro del código o fuera de el (elevar una excepción)

Lectura de un traceback

La manera correcta de leer un traceback es iniciar por el final, en el caso de un error de sintaxis nos indicará en qué línea se encuentra dicho error.
En el caso de excepciones la última línea nos indicará el tipo de exepcion que se generó (generalmente son autoexplicativas pero si no entiendes que paso puedes buscar este error)
La penúltima línea nos indicará dónde se encuentra el error (archivo y línea)
La antepenúltima línea nos muestra “most recent call last” lo que significa que la llamada más reciente es la última (el programa se cerró después de esa llamada, se genero un error)

Elevar una excepción

Cuando tenemos una excepción en python lo que sucede es que se crea un objeto de tipo exception que se va moviendo a través de los bloques de código hasta llegar al bloque principal si es que no se maneja dicha excepción en algún bloque intermedio el programa se interrumpe y genera el traceback

```python
def palindrome(string):
    try:
        if len(string) == 0:
            raise ValueError("No se puede ingresar una cadena vacía")
        return string == string[::-1]
    except ValueError as ve: #le pongo un alias al ValueError
        print(ve)            #imprime el msj de error que seteamos en raise
        return False         #retorna False junto con el error de arriba
    
try:
    print(palindrome(""))
except TypeError:
    print("Sólo se pueden ingresar strings")
```

### Assert statement


Es una manera poco usual de manejar los errores en python

```python
divisors(num):
    divisors = []
    for i in range (1, num + 1):
        divisors.append(i)
    return divisors

def run():
    num = input("Ingresa un número: ")
    assert num.isnumeric(), "Debes ingresar un número" #isnumeric() es un método para strings que pueden ser leidos como numeros
    print(divisors(int(num)))
    print("Terminó mi programa")
    

if __name__ == "__main__":
    run()
```

```python
#reto
def divisor(num):
    divisors = [i for i in range(1,num+1) if num%i == 0]
    return divisors

def run():
    num = input('Enter a number: ')
    assert num.isnumeric() and int(num)>0, 'Enter only positive numbers'
    print(divisor(int(num)))
    print('The program is finished')
```

### Trabajando con archivos de texto en Python


Cuando se quiere leer un archivo en forma de lista abrimos el archivo y le pasamos el parámetro "r" (read). Vamos a leer cada una de las líneas del archivo y cada número o elemento lo agregamos a una lista vacía que creamos al principio
el parámetro encoding = "utf-8" Sirve para que Python pueda soportar caracteres del idioma español, como la “ñ” y letras con tilde. Normalmente al archivo se le nombra con la letra " f " por convención.


Cuando lo que queremos es crear un archivo a partir de una lista de elementos, abrimos el archivo (le ponemos el nombre que queremos que tenga y la ruta donde estará ubicado) con el parámetro "w" (write), este parámetro sobreescribe el archivo, si queremos que agregue otros elementos sin sobresescribir tenemos que usar el parámetro "a" (append). De esta forma aparecerán en el archivo los nombres de cada iteración.
Creamos un ciclo for que recorrerá la lista de nombres e irá escribiendo cada nombre y agregando un salto de página

```python
def read():
    numbers = []
    with open("./archivos/numbers.txt", "r", encoding = "utf-8") as f:
        for line in f:
            numbers.append(int(line))
    print(numbers)

def write():
    names = ["Facundo", "Daniela", "Joaquín", "Agustín"]
    with open("./archivos/names.txt", "w", encoding = "utf-8") as f:
        for name in names:
            f.write(name)
            f.write("\n")


def run():
    #read()
    write()
```

### Atajo en VS Code
Pueden crear un atajo en VS code de la función run y el if para no tener que escribir toda la estructura básica sino que el editor te lo ponga directamente de la siguiente manera:

Abres una nueva terminal en VS code (Terminal-> New Terminal)
Dentro de la consola presionan el shortcut (Ctr + Shift + P)
El anterior comando los lleva a la barra de búsqueda con el símbolo > (en caso de que no se los ponga el editor, lo ponen ustedes), ahí escriben: Snippets
Seleccionan la opción que dice: Preferences: Configure User Snippets
Al seleccionar esa opción les aparecerán todos los lenguajes que soporta VS code y escriben: python y seleccionan la opción que les salga
Al seleccionarla se les abrirá un documento python.json
Dentro del archivo, después de los comentarios que hay, ponen este código:

"Insertar formato de entrada":{
	"prefix": "_run",
	"body": [
		"def run():",
		"    $0",
		"",
		"",
		"",
		"if __name__=='__main__':",
		"    run()"
			
	],
	"description": "Add the entring point"
}

Yo le puse como “prefix”: “_run”, pero ustedes le pueden poner el que deseen, ese va a ser el atajo con el que van a llamar el fragmento de código.

Guardan el archivo
