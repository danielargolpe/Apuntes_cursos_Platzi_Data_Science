#### Datos categricos
Los datos categricos tambin conocidos como datos cualitativos, representan caractersticas como el gnero, el idioma, etc. de una persona. Tambin pueden tomar valores numricos, por ejemplo: 1 para mujeres y 0 para hombres. Hay que tener en cuenta que esos nmeros no tienen significado matemtico.
Los tipos de datos estadsticos categricos se clasifican en:

1. Datos nominales: Otros de los tipos de datos estadsticos son los que tienen valores nominales que representan unidades discretas y se usan para etiquetar variables que no tienen un valor cuantitativo.
Estos datos no tienen un orden, aunque cambiara el orden de sus valores, no cambia su significado.

2. Datos ordinales: Los datos ordinales representan unidades discretas y ordenadas. Por lo tanto, es casi lo mismo que los datos nominales, excepto que su orden es importante.
Las escalas ordinales generalmente, se usan para medir caractersticas no numricas como la felicidad, la satisfaccin del cliente, etc.


#### Datos numricos
Estos tipos de datos estadsticos tambin se conocen como datos cuantitativos, y se refieren a una medida o recuento. Se clasifican de la siguiente manera:

1. Datos discretos: Los datos estadsticos son discretos cuando sus valores son distintos y separados. Es decir, cuando los datos slo pueden tomar ciertos valores.
Este tipo de datos no se puede medir, pero se pueden contar. Bsicamente representan informacin que se puede clasificar.

2. Datos continuos: Los datos continuos representan mediciones y, por lo tanto, sus valores. no se pueden contar, pero se pueden medir. A su vez, estos se clasifican de la siguiente manera:
Datos de intervalo: Los datos de intervalo representan unidades ordenadas que tienen la misma diferencia . Por lo tanto, hablamos de datos de intervalo cuando tenemos una variable que contiene valores numricos que estn ordenados y donde conocemos las diferencias exactas entre los valores.

El problema con los datos de valores de intervalo es que podemos sumar y restar, pero no podemos multiplicar, dividir o calcular razones. Debido a que no existe un cero verdadero, no se pueden aplicar muchas estadsticas descriptivas e inferenciales.

Datos de relacin: Tambin son unidades ordenadas que tienen la misma diferencia. Los datos de relacin son los mismos que los valores de intervalo, con la diferencia de que tienen un cero absoluto

### Medidas de dispersin

Rango: El Rango es el intervalo entre el valor mximo y el valor mnimo.

Cuartiles: Los cuartiles son valores que dividen una muestra de datos en cuatro partes iguales.

1er cuartil (Q1): 25% de los datos es menor que o igual a este valor.

2do cuartil (Q2): La mediana. 50% de los datos es menor que o igual a este valor.

3er cuartil (Q3): 75% de los datos es menor que o igual a este valor.

Rango intercuartil: La distancia entre el primer 1er cuartil y el 3er cuartil (Q3-Q1); de esta manera, abarca el 50% central de los datos.

Diagrama de caja o box plot: representa grficamente una serie de datos numricos a travs de sus cuartiles. De esta manera, el diagrama de caja muestra a simple vista la mediana y los cuartiles de los datos. Tambin puede representar los valores atpicos de estos.

#### Escalamiento o normalizacin lineal
Qu es?
La normalizacin es una tcnica que a menudo se aplica como parte de la preparacin de datos para el aprendizaje automtico.

En qu consiste?
En cambiar los valores de las columnas numricas en el conjunto de datos para usar una escala comn, sin distorsionar las diferencias en los rangos de valores ni perder informacin.

Por qu usarlos?
Los modelos de machine Learning son eficientes en el rango [-1,1]. Si los datos no se encuentran en ese rango debes transformarlos mediante el escalamiento o normalizacin para que los algoritmos modelen los datos correctamente.

Cundo usarlos?
Data simtrica o uniformemente distribuida. Un ejemplo de ellos es la distribucin Gaussiana.
Usando el box-splot y el histograma podremos saber la distribucin de la data.

Tenemos varias opciones para transformar datos numricos:
Cambiar todos los valores a una escala de 0 a 1 o transformar los valores representndolos como rangos de percentiles en lugar de valores absolutos.
Aplicar la normalizacin a una sola columna o a varias columnas en el mismo conjunto de datos.
Si necesita repetir el experimento o aplicar los mismos pasos de normalizacin a otros datos, puede guardar los pasos como una transformacin de normalizacin y aplicarlos a otros conjuntos de datos que tengan el mismo esquema.

Normalizacin lineal
Entre los tipos ms importantes tenemos:

1) Zscore : Dado un conjunto de nmeros calculo su media y desviacin estndar y se asume una distribucin simtrica. Cada dato que tengo (X) lo tengo que convertir a un valor normalizado (Xs). Los valores se transforman mediante la siguiente frmula:
$$z=\frac{x_{}-mean(x)_{}}{stdev(x)_{}}$$

La media y la desviacin estndar se calculan para cada columna por separado. Se utiliza la desviacin estndar de la poblacin.

Luego de la transformacin de los datos la nueva distribucin sigue siendo Gaussiana con media = 0 y desv estandar= 1



2) MinMax : Cada dato que tengo (X) lo tengo que convertir a un valor normalizado (Xs) de ese dato que caiga en el rango [-1, 1] a partir de una transformacin lineal. Los valores se transforman mediante la siguiente frmula:
$$Xs=\frac{2x_{}-min(x)_{}-max(x)_{}}{[max(x)_{}-min(x)_{}]}$$

3) Clipping: Corta la distribucin de los datos entre dos valores limite. El valor de cada dato fuera de los limites colapsa al valor del limite mas cercano. Los valores de min y max se pueden elegir de forma arbitraria o usando los percentiles de manera de agrupar la mayor parte de los datos. Este caso particular se denomina Winzoriding.



Este mtodo omite usar ciertos datos que probablemente funcionen para el modelo aunque sean outliers. En ese sentido, no es muy recomendable usar este mtodo ya que descarta estos datos de la distribucin.

4) Winzoriding: Es un caso particular de Clipping en donde se usan percentiles especficos de limites.

La proyeccin de un vector se obtiene al calcular la "sombra" de ese vector sobre una de sus dos componentes originales x,y. En este ejemplo calculamos la proyeccin del vector sobre la componente x de manera que se anul la componenete en y. Hay algunas direcciones que me van a permitir capturar mejor las varianzas de los datos, en este concepto se basa el anlisis de componentes principales.

Dado un conjunto de datos, cada uno de estos datos tendr un componente en x yotro en y. Si vamos graficando vectores arbitrarios en ese conjunto de datos vemos que cuando la direccin del vector est ms en relacin con la direccin que tienen los datos, las proyecciones son mayores y esto implica una captura de mayor varianza de los datos.
