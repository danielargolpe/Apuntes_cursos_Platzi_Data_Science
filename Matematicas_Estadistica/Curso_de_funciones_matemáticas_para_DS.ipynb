### Funciones polinmicas
Tiene la forma de

$$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una funcin que tiene esta forma se le llama polinomio de grado n. A los elementos a los llamaremos coeficientes donde a  R.

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

```python
N = 100

def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

x = np.linspace(-10,10, num=N)
y = f(x)

plt.plot(x,y)
plt.grid()
```

## Funcin logaritmo

El logaritmo est definido por la **relacin**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que est elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$

```python
def f(x):
    return np.log2(x)

x = np.linspace(0.001,256, num=1000) #definimos un espacio para las x que no contenga al 0 porque el 0 no est definido en la funcin log
y = f(x)

plt.plot(x,y)
```

# Funcin seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funcin escaln de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
    Y = np.zeros(len(x)) # Y temporal, hago un arreglo en donde por la longitud de x las y valgan 0
    for idx, x in enumerate(x): #es una funcin que nos permite obtener el ndice y el arreglo en python, el arreglo que quiero es el de mis x
        if x >= 0:
            Y[idx] = 1
    return Y
   
    
N=1000
x = np.linspace(-10, 10, num=N)
y = H(x)

plt.plot(x,y)
plt.grid()
```

```python
#otra forma de hacerlo con listas

def H(x):

    lista = []

    for i in x:
        if i >= 0:
            i = 1
            lista.append(i)
        else:
            i = 0
            lista.append(i)

    return lista

N = 1000
x = np.linspace(-10,10,num=N)
y = H(x)

plt.plot(x,y)
plt.grid()
```

## Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor que cero, entonces la grfica:

- $y=f(x)+c$ se desplaza $c$ unidades hacia arriba.
- $y=f(x)-c$ se desplaza $c$ unidades hacia abajo.
- $y=f(x-c)$ se desplaza $c$ unidades hacia la derecha.
- $y=f(x+c)$ se desplaza $c$ unidades hacia la izquierda.

```python
N = 1000

def f(x):
  return x**2;

c = 4

x = np.linspace(-10,10, num=N)

y = f(x + c)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

## Alargamientos y compresiones

Siendo $c$ una constante mayor que cero, entonces la grfica:

- $y=c \cdot f(x)$ alarga la grfica verticalmente en un factor de $c$.
- $y= \frac{1}{c} \cdot f(x)$ comprime la grfica verticalmente en un factor de $c$.
- $y=f(c \cdot x)$ comprime la grfica horizontelmente en un factor de $c$.
- $y= f(\frac{1}{c} \cdot x )$ alarga la grfica horizontelmente en un factor de $c$.

```python
N = 1000

def f(x):
  return np.sin(x);

c = 2

x = np.linspace(-15,15, num=N)

y = f((1/c)*x)


fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

### Funcin par
Una funcin es par si cumple la siguiente relacin a lo largo de su dominio:

$f(-x)= f(x)$

Si lo notaste, esta relacin nos dice que una funcin es par si es simtrica al eje vertical (eje Y). Por ejemplo, una parbola es una funcin es par.

### Funcin impar
Una funcin es impar si cumple la siguiente relacin a lo largo de su dominio:

$f(-x)= -f(x)$

Esta relacin nos indica que una funcin es impar si es simtrica al eje horizontal (eje X). Por ejemplo, una funcin cbica es impar.

### Funcin acotada
Una funcin es acotada si su codominio (tambin conocido como rango o imagen) se encuentra entre dos valores, es decir, est acotado. Esta definicin se define como que hay un nmero m que para todo valor del dominio de la funcin se cumple que:

$-m \leq f(x) \leq m $

Por ejemplo, la funcin seno o coseno estn acotadas en el intervalo [-1, 1] dentro de su co-dominio.

### Funciones montonas

Estas funciones son tiles de reconocer o analizar debido a que nos permiten saber si una funcin crece o decrece en alguno de sus intervalos. Que algo sea montono significa que no tiene variaciones. Entonces las funciones montonas son aquellas que dentro de un intervalo I, perteneciente a los nmeros reales, cumple alguna de estas propiedades:

1. La funcin es montona y estrictamente creciente:

Si para todo $x1, x2 \in I: x1 < x2 \Longleftrightarrow f(x1) < f(x2)$


No te preocupes si es un poco difcil de leer esta definicin que para eso estoy aqu para explicarte. La forma correcta de leer esto es si para todo x1 y x2 que pertenecen al intervalo I, tal que x1 sea menor a x2, si y solo si f(x1) sea menor a f(x2). En palabras mucho ms sencillas, lo que nos dice esta definicin es que x1 siempre tiene que ser menor que x2 en nuestro intervalo I, y que al evaluar x2 en la funcin el resultado de esto siempre ser mayor que si evaluamos la funcin en x1. Para las siguientes tres definiciones restantes no cambia mucho la forma en la que se interpretan.

2. La funcin es montona y estrictamente decreciente:

Si para todo $x1, x2 \in I: x1 < x2 \Longleftrightarrow f(x1) > f(x2)$


3. La funcin es montona y creciente:

Si para todo $x1, x2 \in I: x1 < x2 \Longleftrightarrow f(x1) \leq f(x2)$

4. La funcin es montona y decreciente:

Si para todo $x1, x2 \in I: x1 < x2 \Longleftrightarrow f(x1) \geq f(x2)$

### Funciones peridicas

Las funciones peridicas son aquellas que se repiten cada cierto periodo, este periodo se denomina con la letra T. La relacin que debe cumplir la funcin para ser peridica es la siguiente.

$ f(x) = f(x + T), T \neq 0 $

Por ejemplo, la funcin seno y coseno son funciones peridicas con un periodo T = 2. Es decir que si nosotros calculamos f(x) y calculamos f(x + 2) en la funcin seno el valor que nos den ambas expresiones es el mismo.
