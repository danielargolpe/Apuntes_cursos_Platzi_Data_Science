### ¿Para qué se utilizan las siguientes funciones como función de activación?

Función lineal: sirve para mantener valores a lo largo de un proceso, nos devuelve el mismo valor de entrada, por ejemplo predecir el valor de venta

Función escalón o de Heaviside: En esta función cuando entra un valor menor a 0 la función nos retorna 0 y cuando entra un valor mayor o igual a cero la función nos retorna 1. Nos sirve para realizar clasificaciones categóricas, ej. binarias

Función sigmoide: Nos sirve para un proceso de regresión logística, su rango va de 0 a 1, por los que la podemos utilizar para calcular probabilidades

Función tangente hiperbólica: Se la conoce como una función de escalamiento. En esta función los valores van de -1 a 1 y sigue un comportamiento muy similar a la función sigmoide. Desventaja: problemas para aplicarse el logaritmo de backpropagation. 

Función ReLU: Es una función que cuando los valores de entrada son iguales o menores a 0 retorna 0 y cuando son mayores a 0 devuelve x. Nos permite identificar si existe o no un valor para cada neurona en una red o simulación de “neuronas muertas”.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
x = np.linspace(-5,5, num=N)
```

## Función lineal

$$y=mx+b$$

```python
def f(x):
  return x

plt.plot(x, f(x))
plt.grid()
```

## Función escalón o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    

N=1000


y = H(x)

plt.plot(x,y)
plt.grid()
```

## Función sigmoide

$$f(x)=\frac{1}{1+e^{-x}}$$

```python
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

```python
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```

## Función ReLU

$$R(x)=max(0,x)$$

```python
def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x,y)
plt.grid()
```
